# Hidden Pulsars
- rotating neutron stars
- jets of radiation emited by magnetic
- analogous to a light house
- sweeps the eye of the observer i.e. the telescope in this case
- it can have a periodicity
- Periodicity can be as large as millions of years or milliseconds.
- only detectable when the pulse is directly pointing towards the observor.

- improve signal to noise ratio

# Stacking
- take different images,combine into a single image
- compress into 1 single fits image
- average or Mean
- Median: in median frequency of occurence is given importance ,so it is more accurate

```python
%matplotlib inline
import numpy as np
from astropy.io import fits
from matplotlib.colors import LogNorm
import matplotlib.pyplot as plt

data=['/content/sample_data/0001.fits','/content/sample_data/0002.fits','/content/sample_data/0003.fits','/content/sample_data/0004.fits','/content/sample_data/0005.fits']
j=1
for i in data:
  plt.figure(figsize=(4,4))
  plt.title(j)
  j+=1
  plt.imshow(fits.getdata(i))
```

```python
images_append=[]
for i in data:
  images_append.append(fits.getdata(i))
print(images_append)
print(sum(images_append)/5)
plt.imshow(sum(images_append)/5)
```
