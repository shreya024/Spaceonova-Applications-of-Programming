### Q1. On the given set of FITS files, perform MEAN and MEDIAN stacking and write down the differences you notice

```python
%matplotlib inline
import numpy as np
from astropy.io import fits
from matplotlib.colors import LogNorm
import matplotlib.pyplot as plt
```

```python
# The code for stacking
image_concat=[]
for i in range(100):
  a=fits.getdata('/content/sample_data/pulsar_100/{0:04}.fits'.format(i))
  image_concat.append(a)
stack_mean=np.mean(image_concat,axis=0)
stack_median=np.median(image_concat,axis=0)
```

```python
plt.figure(figsize=(6,6))
plt.imshow(fits.getdata('/content/sample_data/pulsar_100/0040.fits'))
plt.colorbar()
plt.title('Random Image')

plt.figure(figsize=(6,6))
plt.imshow(stack_mean)
plt.colorbar()
plt.title('Mean')

plt.figure(figsize=(6,6))
plt.imshow(stack_median)
plt.colorbar()
plt.title('Median')
```

```python
# The observations you can conclude
```

- In median frequency of occurence is given importance ,so it is more accurate.
- when some of the values are more extreme, the effect on the median is smaller as compared to mean.
- Here,pulsars are hardly visible in mean and random image
- However,it is distinguishable in the case of median
